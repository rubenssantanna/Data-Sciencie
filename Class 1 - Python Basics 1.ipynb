## O que é programação em Python?

A programação é o **processo de criação de um conjunto de instruções que dizem a um computador como realizar uma tarefa.** Um programa pode ser escrito usando uma variedade de "linguagens" de programação como Java, Python, C, C ++ etc.

Essencialmente, a programação é o processo de comunicação com um computador através de uma linguagem que o computador entende. Para poder dirigir efetivamente um computador para fazer o que precisamos, aprendemos a traduzir nossos requisitos para esta linguagem como código/rotinas que ele pode executar repetidamente. Estaremos usando a linguagem Python!

## Syntaxe em Python

Uma linguagem de programação, inclui muitos elementos que são análogos aos elementos que compõem uma linguagem humana como o Português. Por exemplo, a linguagem de programação possui:

- variáveis, constantes e palavras-chave (também chamadas de palavras reservadas) 
- semelhantes às palavras inglesas

- programas - semelhantes aos parágrafos

### Regras para nomes de variáveis

Para garantir a sintaxe correta, as regras devem ser respeitadas ao atribuir nomes de variáveis; Estes podem ser nomes curtos (como x ou y) ou nomes mais descritivos (como my_text, ou volume, etc.).

Regras para a definição de variáveis em Python:
- o nome de uma variável deve **começar com uma letra ou o símbolo -**
- o nome de uma variável **Não pode começar com um número***
- o nome de uma variável **só pode conter caracteres alfanuméricos e sublinhados (A-z, 0-9, e _)**
- o nome de uma variável diferencia **letras maiúsculas e minúsculas***! (as variáveis Idade, idade e IDADE são 3 variáveis diferentes)

Há também **nomes reservados que não podem ser usados para nomear variáveis**, uma vez que eles têm sua própria função em Python.

## Tipos de dados e variáveis em Python

### Introdução aos tipos de dados

Ao programar é importante entender o tipo de dados com os quais estamos lidando.

Como já discutimos, as variáveis podem conter diferentes tipos de dados, e diferentes tipos de dados podem fazer diferentes tipos de coisas.

Os tipos de dados nativos do Python são:
- números: inteiros (int), decimais (float) e números complexos (complex) 
- texto (str)
- listas (list)
- tuples (tuple)
- conjuntos (set)
- dicionários (dict)

Cada tipo de dado é manipulado utilizando:
- operadores
- funções
- métodos do tipo de dados

#### Determinando o tipo de dados de uma variável

Utilizando a função:

**type()**

aplicado a uma variável, obtemos o tipo de dados que a variável armazena!

##### Exemplos

Obtemos que o tipo de dados na variável x são números inteiros **(int)**

```python
x = 5 
print(type(x))
```

Ou seja, a variável x é de classe números inteiros

#### Casting(Moldagem): Definição do tipo de dados de uma variável

Embora em Python não seja necessário definir explicitamente o tipo de dados de uma variável, já que a variável herda o formato do valor que armazena, haverá momentos em que desejamos especificar o tipo da variável.


Isto pode ser feito através do que chamamos de moldagem(casting).

Como Python é uma linguagem orientada a objetos, para definir tipos de dados, as classes são utilizadas através de funções de construtor:

**- int ()** - construir um número inteiro a partir de um número inteiro, um float (arredondamento para baixo) ou um número textual

**- float ()** - construir um número float a partir de um número inteiro, um número float ou um número textual

**- str ()** - construir uma cadeia de texto a partir de um texto ou números

##### Exemplo: **int**()

```python
x = int(1) # x será 1
y = int(2.8) # y será 2
z = int("3") # z será 3
```

```python
print(x,type(x))
print(y,type(y))
print(z,type(z))
```

##### Exemplo: **float()**

```python
x = float(1)     # x será 1.0
y = float(2.8)   # y será 2.8
z = float("3")   # z será 3.0
w = float("4.2") # w será 4.2
```

```python
print(x,type(x))
print(y,type(y))
print(z,type(z))
print(w,type(w))
```

##### Exemplo: **str()**

```python
x = str("s1") # x será 's1'
y = str(2)    # y será '2'
z = str(3.0)  # z será '3.0'
```

```python
print(x,type(x))
print(y,type(y))
print(z,type(z))
```

### Variáveis numéricas

#### Tipos de dados numéricos

Há 3 tipos de dados numéricos em Python:

**- inteiros (int):** são inteiros, positivos ou negativos, sem decimais e sem um comprimento fixo

**- 7 dígitos(32-bit) (float):** número positivo ou negativo que contém decimais

#### Conversão entre tipos numéricos

Podemos transformar uma variável numérica entre os 3 tipos de dados usando os métodos **int()**, **float()** ou **complex()**.

- **o método int()** converte uma variável float em int
- **o método float()** converte uma variável int para float
- **o método complex()** converte uma variável int ou float em complexa

**Exceção!** Uma variável complexa não pode ser lançada para int ou float.
