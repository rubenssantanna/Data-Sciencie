```python
import numpy as np
import pandas as pd
```

```python
url = 'https://drive.google.com/file/d/1t8BVy12jkBoVFBhkpkS28hYesKiNb1kQ/view?usp=sharing'
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]
df_titanic = pd.read_csv(url)

dimension = df_titanic.shape
number_of_elements = len(df_titanic)
names_of_columns = list(df_titanic.columns)
names_of_rows = list(df_titanic.index)
types_of_columns = df_titanic.dtypes
first_10_rows = df_titanic.head(10)
last_10_rows = df_titanic.tail(10)
```

```python
#Mostrar na tela os nomes das pessoas que estavam na primeira classe em ordem alfab√©tica.

df_titanic[df_titanic["Pclass"] == 1].sort_values(by='Name')
```

```python
#Mostrar na tela a porcentagem de pessoas que sobreviveram e morreram.

survived = len(df_titanic[df_titanic["Survived"] == 1])
did_not_survive = len(df_titanic[df_titanic["Survived"] == 0])
percentage_survived = round((survived / number_of_elements) * 100, 2)
percentage_did_not_survive = round((did_not_survive / number_of_elements) * 100, 2)
print('Survived({}):'.format(survived), str(percentage_survived) + '%,', 'Did not survive({}):'.format(did_not_survive), str(percentage_did_not_survive) + '%')
```

```python
#Mostrar na tela a porcentagem de pessoas que sobreviveram em cada classe.

classesGroup = df_titanic.groupby(by='Pclass')

print('Percentage of survived by class:')
for name, group in classesGroup:
    number_of_peoples = len(group)
    survived = len(group[group.Survived == 1])
    did_not_survive = len(group[group.Survived == 0])
    percentage_survived = round((survived / number_of_peoples) * 100, 2)
    percentage_did_not_survive = round((did_not_survive / number_of_peoples) * 100, 2) 
    print(str(name) + '¬∞ Class ({}) ->'.format(number_of_peoples), 'Survived({}):'.format(survived), str(percentage_survived) + '%,', 'Did not survive({}):'.format(did_not_survive), str(percentage_did_not_survive) + '%')
```

```python
#Remover passageiros com idade desconhecida da DataFrame.

df_titanic = df_titanic.dropna(subset=['Age'])
display(df_titanic)
```

```python
classesGroup = df_titanic[df_titanic['Sex'] == 'female'].groupby(by='Pclass')

print('Mean age of woman by class:')
for name, group in classesGroup: 
    mean = int(group.agg({'Age':'mean'}).Age)
    print('{}¬∞ Class -> {}'.format(name, mean))
```

```python
df_titanic = df_titanic.assign(Under_Age = df_titanic['Age'] < 18)
display(df_titanic)
```

```python
#Mostrar na tela a porcentagem de menores e adultos que sobreviveram em cada classe.

classesGroup = df_titanic.groupby(by='Pclass')

print('Percentage of under age and adult:')
for name, group in classesGroup:
    total_peoples = len(group)
    number_of_under_age = len(group[group['Under_Age']])
    number_of_adult = total_peoples - number_of_under_age
    percentage_under_age = round((number_of_under_age / total_peoples) * 100, 2)
    percentage_adult = round((number_of_adult / total_peoples) * 100, 2) 
    print('{}¬∞ Class -> Under Age: {}%, Adult: {}%'.format(name, percentage_under_age, percentage_adult))
```
